```
#dataframe oluştur
import pandas as pd
datas={
    "isim":["Banu","İnanç","Şengül","Orhan","Aysuhan","Volkan","Şahin","Alican","Selin"],
    "yas":[44,46,56,62,42,40,35,34,32],
    "maas":[86000,95000,88000,96000,60000,75000,55000,52000,40000],
    "alan":["matematiksel fizik","kuantum bilişimi","teorik mekanik","yüksek enerji","manyetizma","programlama","katıhal fiizği","optik","atom ve molekül"]
}
df=pd.DataFrame(datas)
df
```

```
#ilk 3 satırı göster
df.head(3)
```

```
#yaşı 40tan büyük olanları göster
greater=df[df["yas"]>40]
greater
```

```
#maasların %10 zamlı halini gösteren yeni sütun ekle
df["yeni maas"]=df["maas"]+df["maas"]*0.1
df
```

```
#maaslara göre azan olarak sırala
df.sort_values("maas")
```

```
#küçükten büyüğe sırala
df.sort_values(by="maas",ascending=False)
```

```
#alana göre gruplandırıp maasların ortalamasını al
df.loc[len(df)]=["Emre",26,35000,"manyetizma",38500]
df
```

```
df.groupby("alan")["maas"].mean()
```
